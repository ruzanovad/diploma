```python
file_paths = [
    "/home/kaiser/Downloads/tb_logs_image2latex_model_version_13.csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_version_14.csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_from_kaggle_3.csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_from_kaggle_2.csv"
]

# Load and merge CSV files on common structure, tagging them for identification
dfs = []
for path in file_paths:
    df = pd.read_csv(path)
    # df['source'] = os.path.basename(path)  # add a source column to identify the origin
    dfs.append(df)

val_bleu = pd.concat(dfs, ignore_index=True).sort_values(by=['Step'], ascending=True)
```

```python
import matplotlib.pyplot as plt

last_x = val_bleu['Step'].iloc[-1]
last_y = val_bleu['Value'].iloc[-1]

plt.title("BLEU на валидационной выборке")

plt.scatter(last_x, last_y, color='red')
plt.text(last_x-0.1,  last_y-0.1, f'{last_y:.4f}', 
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
         ha='left', va='bottom')
plt.plot(val_bleu['Step'], val_bleu['Value'])
```

```python
file_paths = [
    "/home/kaiser/Downloads/tb_logs_image2latex_model_version_13 (2).csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_version_14 (2).csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_from_kaggle_3 (2).csv",
    "/home/kaiser/Downloads/tb_logs_image2latex_model_from_kaggle_2 (2).csv"
]

# Load and merge CSV files on common structure, tagging them for identification
dfs = []
for path in file_paths:
    df = pd.read_csv(path)
    # df['source'] = os.path.basename(path)  # add a source column to identify the origin
    dfs.append(df)

em = pd.concat(dfs, ignore_index=True).sort_values(by=['Step'], ascending=True)
em
```

```python
plt.title("Доля точных совпадений предсказанного и истинного текстов на валидационной выборке")
last_x = em['Step'].iloc[-1]
last_y = em['Value'].iloc[-1]


plt.scatter(last_x, last_y, color='red')
plt.text(last_x,  last_y-0.02, f'{last_y:.4f}', 
         bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
         ha='left', va='bottom')
plt.plot(em['Step'], em['Value'])
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharex=True)

# BLEU
axs[0].plot(val_bleu['Step'], val_bleu['Value'])
last_x = val_bleu['Step'].iloc[-1]
last_y = val_bleu['Value'].iloc[-1]
axs[0].scatter(last_x, last_y, color='red')
axs[0].text(last_x - 0.1, last_y - 0.05, f'{last_y:.4f}',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
            ha='left', va='bottom')
axs[0].set_title("BLEU на валидационной выборке")
axs[0].set_xlabel("Step")
axs[0].grid(True)

# Levenshtein
axs[1].plot(edit_distance['Step'], edit_distance['Value'])
last_x = edit_distance['Step'].iloc[-1]
last_y = edit_distance['Value'].iloc[-1]
axs[1].scatter(last_x, last_y, color='red')
axs[1].text(last_x, last_y + 0.1, f'{last_y:.4f}',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
            ha='left', va='bottom')
axs[1].set_title("Нормированное среднее расстояние Левенштейна\n на валидационной выборке")
axs[1].set_xlabel("Step")
axs[1].grid(True)

# Exact Match
axs[2].plot(em['Step'], em['Value'])
last_x = em['Step'].iloc[-1]
last_y = em['Value'].iloc[-1]
axs[2].scatter(last_x, last_y, color='red')
axs[2].text(last_x, last_y - 0.02, f'{last_y:.4f}',
            bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'),
            ha='left', va='bottom')
axs[2].set_title("Доля точных совпадений предсказанного и истинного \n текстов на валидационной выборке")
axs[2].set_xlabel("Step")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
