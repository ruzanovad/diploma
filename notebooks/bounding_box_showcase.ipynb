```python
import cv2
import torch
import numpy as np
import torchvision.transforms as T
from torchvision.utils import draw_bounding_boxes
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def test(
    image="../datasets/number/dataset/images/train/3_.png",
    template="../datasets/number/patterns/10/0.png",
    method=cv2.TM_CCOEFF_NORMED,
    threshold=0.8,
):
    # Load an image using PIL
    imagee = Image.open(image).convert("RGB")

    image_tensor = T.ToTensor()(imagee) * 255

    # Загрузим изображение и шаблон
    image = cv2.imread(image, 0)
    template = cv2.imread(template, 0)

    # Размер шаблона

    w, h = template.shape[::-1]
    # Выполним сопоставление
    res = cv2.matchTemplate(image, template, method)

    # Установим порог обнаружения
    loc = np.where(res >= threshold)

    # Обведём найденные участки
    for pt in zip(*loc[::-1]):
        print(pt)
        print((pt[0] + w, pt[1] + h))
    bbox = torch.tensor(
        [[pt[0], pt[1], pt[0] + w, pt[1] + h] for pt in zip(*loc[::-1])]
    )  # Example rectangle coordinates

    # Define color and label
    # colors = ["red", "blue", "green", "yellow", "purple", "orange"]
    # labels = ["Detected Object"]

    # Draw bounding box on the image
    image_with_bbox = draw_bounding_boxes(
        image_tensor.byte(),
        bbox,
        # labels=labels,
        colors="red",
        width=3,
    )

    # Convert to PIL image for visualization
    image_with_bbox_pil = T.ToPILImage()(image_with_bbox)

    # Show the image
    plt.imshow(image_with_bbox_pil)
    plt.axis("off")
    plt.show()
```

```python
test()
```

```python
test(method=cv2.TM_CCOEFF)
```

```python
test(threshold=0.95)
```
