```python
from ultralytics import YOLO
import yaml
```

```python
best_model = YOLO("../latex/yolo11pt_2025-05-20_12-02-37/weights/best.pt")
```

```python
best_model
```

```python
# Load config (Hydra-style YAML)
with open("../configs/yolo.yaml", "r") as f:
    cfg = yaml.safe_load(f)


metrics = best_model.val(
    data="../"+cfg["data"], 
    imgsz=cfg["imgsz"], 
    batch=cfg["batch"],     
    device=cfg["device"],  
    split="test", 
    save=cfg.get("visualize", False), 
    plots=cfg.get("plots", False),
    save_txt=True
)
```
