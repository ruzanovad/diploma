```python
from config import get_config
from pathlib import Path
import torch
import torch.nn as nn
from config import latest_weights_file_path
from train import get_model, get_ds, run_validation
from translate import translate

cfg = get_config()
# cfg['model_folder'] = '..//drive/MyDrive/Models/pytorch-transformer/weights'
# cfg['tokenizer_file'] = '..//drive/MyDrive/Models/pytorch-transformer/vocab/tokenizer_{0}.json'

cfg['model_folder'] = '/home/kaiser/Downloads/'
cfg['tokenizer_file'] = '/home/kaiser/Downloads/tokenizer_{0}.json'

cfg['datasource'] = "Helsinki-NLP/opus_books"
cfg['batch_size'] = 36
cfg['num_epochs'] = 100
# cfg['preload'] = None
cfg['seq'] = 256
```

```python
# Define the device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
train_dataloader, val_dataloader, tokenizer_src, tokenizer_tgt = get_ds(cfg)
model = get_model(cfg, tokenizer_src.get_vocab_size(), tokenizer_tgt.get_vocab_size()).to(device)

# Load the pretrained weights
model_filename = latest_weights_file_path(cfg)
state = torch.load(model_filename, map_location=torch.device('cpu'))
model.load_state_dict(state['model_state_dict'])
```

```python
run_validation(model, val_dataloader, tokenizer_src, tokenizer_tgt, cfg['seq_len'], device, lambda msg: print(msg), 0, None, num_examples=10)
```
